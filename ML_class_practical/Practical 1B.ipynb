```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error

# past data for training the regression model

# Example: x is CGPA and  y is Salary
x=np.array([6.1, 6.15, 6.30, 7.24, 7.50, 7.50, 7.9, 8.0, 8.9, 9.1, 9.5, 9.5, 9.52]).reshape(-1,1)
y=np.array([0.0, 2.50, 2.25, 6.00, 3.30, 3.75, 4.5, 3.3, 4.0, 3.5, 6.5, 10.5, 14.5]).reshape(-1,1)

n=len(x) #sample size
#plot the training instances(points)
plt.scatter(x,y,color='blue')

#Defining and fitting the model
ols=linear_model.LinearRegression()
ols.fit(x,y)

#Visualizing the linear model
X = np.linspace(0,10, 50, endpoint=True).reshape(-1,1)
for i in range(50):
    Yp=ols.predict(X)

xl="CGPA"
yl="Salary"
plt.xlabel(xl) 
plt.ylabel(yl)
plt.plot(X, Yp)


#For the given CGPA find the  predicted salary
cgpa=np.array([7.2]).reshape(-1,1)
psal=ols.predict(cgpa)
plt.plot(cgpa[0],psal[0],'ro')
print('Predicted Salary for the student with CGPA','%.2f'%cgpa[0],' is Rs','%.2f'%psal[0], 'Lac')

#Computing R square and MSE values
yp=ols.predict(x)
r2=round(r2_score(y,yp),2)
plt.title("Linear Ordinary LS fit Model - R2 Score="+str(r2))
mse=mean_squared_error(y,yp)
print('MSE=','%.2f'%mse,'R2 value=','%.2f'%r2)
```
